3. Реалізація алгоритму (Перевірка паліндрома)

```python
import re

def check_palindrome_variant_7(text):
    # Контроль правильності введення даних
    if not isinstance(text, str):
        raise TypeError("Помилка: Потрібно ввести рядок (текст).")

    # Очищення рядка: видаляємо пробіли, розділові знаки та переводимо в нижній регістр
    # Додано українські літери для коректної перевірки україномовних паліндромів
    cleaned = re.sub(r'[^а-яА-Яa-zA-Z0-9іІїЇєЄґҐ]', '', text).lower()

    if len(cleaned) == 0:
        return "Помилка: рядок не містить літер або цифр для перевірки."

    # Перевірка на паліндром
    if cleaned == cleaned[::-1]:
        return f'Рядок "{text}" є паліндромом.'
    else:
        return f'Рядок "{text}" не є паліндромом.'
```

4. Приклади

```python
# Тестові приклади
print("--- Перевірка ---")
print(check_palindrome_variant_7("Падав сніг на поріг"))
print(check_palindrome_variant_7("Rotor"))
print(check_palindrome_variant_7("Іспанія"))

# Приклад контролю введення
print("\n--- Перевірка помилки ---")
try:
    print(check_palindrome_variant_7(98765))
except Exception as e:
    print(e)
```

```python
import matplotlib.pyplot as plt

# Список для тестування (як паліндроми, так і звичайні слова)
words_to_test = ["Rotor", "Pylyp", "Іспанія", "Падав сніг на поріг", "Madrid", "level"]
results_binary = []

for item in words_to_test:
    cleaned = re.sub(r'[^а-яА-Яa-zA-Z0-9іІїЇєЄґҐ]', '', item).lower()
    # 1 - якщо паліндром, 0 - якщо ні
    results_binary.append(1 if cleaned == cleaned[::-1] else -1)

# Побудова графіка
plt.figure(figsize=(10, 5))
# Паліндроми будуть зеленими, не паліндроми - червоними
colors = ['green' if res == 1 else 'red' for res in results_binary]

plt.bar(words_to_test, results_binary, color=colors)
plt.title("Перевірка рядків на паліндром (Варіант 7)")
plt.ylabel("Результат (1 - Так, 0 - Ні)")
plt.xticks(rotation=45)
plt.yticks([0, 1], ['Ні', 'Так'])
plt.tight_layout()
plt.show()
```
